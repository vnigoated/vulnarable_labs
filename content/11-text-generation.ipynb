# Lab 11: Text Generation (N-Gram)

## Introduction
How does your phone suggest the next word? One simple method is **N-Grams** - looking at what word usually follows the previous one.

## Goal
Build a simple auto-complete function.

```python
import random

text = "i like apples i like bananas i like coding i love python i love ai"
tokens = text.split()

# Build Markov Chain
chain = {}

for i in range(len(tokens) - 1):
    current_word = tokens[i]
    next_word = tokens[i+1]
    
    if current_word not in chain:
        chain[current_word] = []
    chain[current_word].append(next_word)
    
print("Model:", chain)

def predict_next(word):
    if word in chain:
        return random.choice(chain[word])
    return "?"

print("Next word after 'i':", predict_next('i'))
print("Next word after 'like':", predict_next('like'))
```
