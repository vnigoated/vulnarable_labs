# Lab 9: Project - Resume Parser

## Goal
Build a tool that takes a resume (text) and creates a structured profile (Email, Phone, Skills).

## Strategy
1.  Use **Regex** for Phones and Emails.
2.  Use **Keyword Matching** for Skills (Python, Java, etc.).

```python
import re

resume_text = """
John Doe
Software Engineer
Contact: john.doe23@gmail.com | (555) 123-4567
Skills: Proficient in Python, Java, SQL, and Machine Learning.
"""

def parse_resume(text):
    profile = {}
    
    # Email
    email = re.search(r'[\w\.-]+@[\w\.-]+', text)
    profile['email'] = email.group(0) if email else None
    
    # Phone
    phone = re.search(r'\(?\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}', text)
    profile['phone'] = phone.group(0) if phone else None
    
    # Skills
    known_skills = ['Python', 'Java', 'C++', 'SQL', 'React', 'Machine Learning']
    found_skills = [s for s in known_skills if s.lower() in text.lower()]
    profile['skills'] = found_skills
    
    return profile

print(parse_resume(resume_text))
```
