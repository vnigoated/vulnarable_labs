We will build a Naive Bayes Spam classifier and then use it interactively.

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Simple Dataset
data = {
    'message': ['Call me later', 'Win a free iPhone now!', 'Hello friend', 
                'URGENT! You have won cash.', 'Let us meet for lunch', 'Free entry in contest'],
    'label': ['ham', 'spam', 'ham', 'spam', 'ham', 'spam']
}
df = pd.DataFrame(data)

# Train Model
model = Pipeline([
    ('vect', CountVectorizer()),
    ('clf', MultinomialNB()),
])
model.fit(df['message'], df['label'])
print("Model Trained!")
```

## Interactive Spam Tester
Type a message and check if it's Spam.

```python
import ipywidgets as widgets

def check_spam(message):
    prediction = model.predict([message])[0]
    proba = model.predict_proba([message]).max() * 100
    
    color = "red" if prediction == 'spam' else "green"
    print(f"\033[1;3%dmPrediction: {prediction.upper()} ({proba:.1f}%% confidence)\033[0m" % (1 if prediction=='spam' else 2))

input_box = widgets.Text(placeholder='Type a message...', description='Message:')
widgets.interactive(check_spam, message=input_box)
```
