# Lab 7: Sentiment Analysis

## Introduction
Sentiment analysis determines the emotional tone behind words. Is a review Positive, Negative, or Neutral?
We will use **TextBlob**, a simple library built on NLTK.

```python
%pip install textblob
```

```python
from textblob import TextBlob

text = "I absolutely love this new phone! It is amazing."
blob = TextBlob(text)

# Polarity: -1 (Negative) to +1 (Positive)
# Subjectivity: 0 (Objective) to 1 (Subjective)
print(f"Text: {text}")
print(f"Polarity: {blob.sentiment.polarity}") # Should be positive
print(f"Subjectivity: {blob.sentiment.subjectivity}")
```

## Interactive Analyzer
