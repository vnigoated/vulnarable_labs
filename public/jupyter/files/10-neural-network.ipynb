## Introduction
Before using PyTorch or TensorFlow, let's understand how a neuron works by building one from scratch using `numpy`.

## The Perceptron
A Perceptron takes inputs, multiplies them by weights, adds a bias, and passes it through an activation function.

```python
import numpy as np

class Perceptron:
    def __init__(self, input_size):
        self.weights = np.zeros(input_size)
        self.bias = 0
        self.lr = 0.1
        
    def activation(self, x):
        # Step function
        return 1 if x >= 0 else 0
    
    def predict(self, x):
        z = np.dot(self.weights, x) + self.bias
        return self.activation(z)
    
    def train(self, X, y, epochs=10):
        for _ in range(epochs):
            for i in range(len(X)):
                prediction = self.predict(X[i])
                error = y[i] - prediction
                
                # Update weights
                self.weights += self.lr * error * X[i]
                self.bias += self.lr * error

# Logic Gate (OR function)
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([0, 1, 1, 1])

p = Perceptron(input_size=2)
p.train(X, y)

print("Prediction for [0, 1]:", p.predict([0, 1]))
```
