# Lab 8: Named Entity Recognition (NER)

## Introduction
NER involves extracting real-world objects like Persons (PER), Locations (LOC), and Organizations (ORG) from text.
We will use NLTK's `ne_chunk` for this, as it is lightweight enough for the browser.

```python
import nltk
import micropip
await micropip.install('nltk')

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
text = "Apple Inc. was founded by Steve Jobs in California."

# 1. Tokenize
tokens = nltk.word_tokenize(text)

# 2. Part of Speech Tagging
pos_tags = nltk.pos_tag(tokens)

# 3. Chunking (NER)
chunks = nltk.ne_chunk(pos_tags)

print("Detected Entities:")
for chunk in chunks:
    if hasattr(chunk, 'label'):
        print(f"{chunk.label()}: {' '.join(c[0] for c in chunk)}")
```
